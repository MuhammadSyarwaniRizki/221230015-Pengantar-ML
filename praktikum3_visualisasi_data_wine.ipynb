```python
# === Import Library ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

# === Load Dataset ===
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df["target"] = wine.target

# === Pilih 3 fitur paling membedakan antar kelas ===
features = ["flavanoids", "proline", "alcohol"]

# === Visualisasi Gabungan Box + Violin Plot ===
plt.figure(figsize=(15, 12))

for i, col in enumerate(features, 1):
    plt.subplot(2, 2, i)
    # Violin plot: menunjukkan distribusi data
    sns.violinplot(x="target", y=col, data=df, palette="Pastel1", inner=None)
    # Box plot: menunjukkan median dan IQR
    sns.boxplot(x="target", y=col, data=df, width=0.2, boxprops={'zorder': 2}, showfliers=False)

    plt.title(f"{col.capitalize()} per Kelas Wine", fontsize=14)
    plt.xlabel("Kelas Wine (0, 1, 2)")
    plt.ylabel(col.capitalize())

plt.tight_layout()
plt.show()
```

```python
# === LOAD DATASET WINE ===
import pandas as pd
from sklearn.datasets import load_wine

data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

# Cek apakah df sudah ada
print(df.head())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

#PCA PROJECTION

# Pisahkan fitur dan target
X = df.drop("target", axis=1)
y = df["target"]

# PCA ke 2 dimensi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Buat dataframe untuk visualisasi
df_pca = pd.DataFrame({
    "PC1": X_pca[:,0],
    "PC2": X_pca[:,1],
    "target": y
})

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="target", palette="Set2", s=70, alpha=0.8)
plt.title("PCA Projection: Overlap antar Kelas Wine", fontsize=14)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA

#PCA PROJECTION

# Load dataset wine
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

# Pisahkan fitur dan target
X = df.drop("target", axis=1)
y = df["target"]

# PCA ke 2 dimensi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# DataFrame hasil PCA
df_pca = pd.DataFrame({
    "PC1": X_pca[:,0],
    "PC2": X_pca[:,1],
    "target": y
})

# Visualisasi PCA projection
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="target", palette="Set1", s=70, alpha=0.8)
plt.title("Visualisasi Multi-Kelas dengan PCA Projection", fontsize=14)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA

# Load dataset wine
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

# Pisahkan fitur dan target
X = df.drop("target", axis=1)
y = df["target"]

# PCA ke 2 dimensi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# DataFrame hasil PCA
df_pca = pd.DataFrame({
    "PC1": X_pca[:,0],
    "PC2": X_pca[:,1],
    "target": y
})

# Visualisasi PCA
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="target", palette="Set1", s=70, alpha=0.8)
plt.title("Kelompok Wine Berdasarkan Karakteristik Kimia (PCA Projection)", fontsize=14)
plt.show()
```
